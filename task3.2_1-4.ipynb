4. Задание (в программе)
Решите систему уравнений и неравенств:
    
A.
$$
exp(x) + x∙(1 – y) = 1\\  
y = x2 – 1
$$
B.
$$
exp(x) + x∙(1 – y)  - 1 > 0\\
y = x2 – 1
$$

A.

```python
from scipy.optimize import fsolve

 
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) - 1)/x +1)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10,10) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return ( x**2 - 1-y, (np.exp(x) - 1)/x +1-y)
x1, y1 =  fsolve(equations, (-1, 1))
x2, y2 =  fsolve(equations, (2, 5))
              
print (f"x1y1[{x1:.1f}; {y1:.1f}]")
print (f"x2y2[{x2:.1f}; {y2:.1f}]")
```

B.

```python
x = np.linspace(0, 10, 201)
plt.plot(x, (np.exp(x) - 1)/x +1)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0,10) 
plt.xlim(0,4)
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return ( x**2 - 1-y, (np.exp(x) - 1)/x +1-y)

x, y =  fsolve(equations, (2, 5))
              
print (f"[{x:.1f}; {y:.1f}]")
```
